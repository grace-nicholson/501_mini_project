```python
import matplotlib as mpl
import pandas as pd
mpl.rcParams['axes.formatter.useoffset'] = False    # Remove annoying auto-offset
import matplotlib.pyplot as plt
import numpy as np

# Read in the data from a CSV file
df=pd.read_csv(r"/Users/grace/Documents/Masters/501 project/501ProjectEXP3-4.csv")

# add time in minutes column
df['time_minutes']= df['cycles']/2 # convert cycles to minutes

# statistic filtering to remove outliers
lower_bound = df['air quality'].quantile(0.08)
upper_bound = df['air quality'].quantile(0.99)

airq_clean = df[(df['air quality'] > lower_bound) & (df['air quality'] < upper_bound)]

# Create a figure
fig = plt.figure(figsize=(10, 6))
# Add an axis to the figure
ax = fig.add_subplot(111)

# Tell the axis to plot the data - time on the x-axis, temperature on the y-axis
ax.scatter(airq_clean['time_minutes'], airq_clean['temp'])

# Tell the axis to plot the data - time on the x-axis, air quality reading on the y-axis, with a different colour
ax.scatter(airq_clean['time_minutes'], airq_clean['air quality'], c=(0 , 0.5, 0))

# Tell the axis to plot the data - time on the x-axis, humidity reading on the y-axis, with a different colour
ax.scatter(airq_clean['time_minutes'], airq_clean['humidity'], c=(0.8 , 0, 0.2))

# Set labels, title, legend, and axis limits
ax.set_xlabel('Time (minutes)', fontweight='bold', fontsize=14)
ax.set_ylabel('air quality (PPM)', fontweight='bold', fontsize=14)
ax.set_title('Temperature/Air quality/Humidity vs. Time', fontweight='bold', fontsize=14)
plt.legend(['Temperature (C)','Air quality', 'Humidity (%)'])
ax.set_xlim(left=0)
ax.set_ylim(ymin=0,ymax=200)

# Create a second y-axis to plot humidity and temperature
ax2 = ax.twinx()
ax2.set_ylabel('humidity (%)/ temp (C)', fontweight='bold', fontsize=14)
ax2.set_ylim(ymin=20,ymax=30)


# Show the figure
plt.show()
```
