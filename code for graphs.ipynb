```python
import matplotlib as mpl
import pandas as pd
mpl.rcParams['axes.formatter.useoffset'] = False    # Remove annoying auto-offset
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv(r"/Users/grace/Documents/Masters/501 project/501ProjectEXP2-4.csv")

#df['air quality'] = pd.to_numeric(df['air quality'], errors='coerce')
#df['humidity'] = pd.to_numeric(df['humidity'], errors='coerce')
#df['temp'] = pd.to_numeric(df['temp'], errors='coerce')

df['time_minutes']= df['cycles']/2 # convert cycles to minutes

lower_bound = df['air quality'].quantile(0.06)
upper_bound = df['air quality'].quantile(0.99)

# Here's the tricky bit - we ignore values that are below the lower bound or above the upper bound - ask ChatGPT what the code is doing!
airq_clean = df[(df['air quality'] > lower_bound) & (df['air quality'] < upper_bound)]

# Create a figure
fig = plt.figure(figsize=(10, 6))
# Add an axis to the figure
ax = fig.add_subplot(111)
# Tell the axis to plot the data - time on the x-axis, temperature on the y-axis
ax.scatter(airq_clean['time_minutes'], airq_clean['temp'])

# Tell the axis to plot the data - time on the x-axis, air quality reading on the y-axis
ax.scatter(airq_clean['time_minutes'], airq_clean['air quality'], c=(0 , 0.5, 0))

# Tell the axis to plot the data - time on the x-axis, humidity reading on the y-axis
ax.scatter(airq_clean['time_minutes'], airq_clean['humidity'], c=(0.8 , 0, 0.2))

ax.set_xlabel('Time (minutes)', fontweight='bold', fontsize=14)
# Bonus question - search online to find out what the funny r and $^\circ$ are doing in the line below
ax.set_ylabel('air quality (PPM)', fontweight='bold', fontsize=14)
ax.set_title('Temperature/Air quality/Humidity vs. Time', fontweight='bold', fontsize=14)
plt.legend(['Temperature (C)','Air quality', 'Humidity (%)'])
xmin, xmax = ax.get_xlim()
#ymin, ymax = ax.get_ylim()
ax.set_xlim(left=0)
ax.set_ylim(ymin=0,ymax=300)

ax2 = ax.twinx()
ax2.set_ylabel('humidity (%)/ temp (C)')
ax2.set_ylim(ymin=20,ymax=30)


# Show the figure - not strictly necessary in jupyter notebook, but essential if you're writing a .py file
plt.show()
```
